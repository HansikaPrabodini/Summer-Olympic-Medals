### Z-Score Normalization 

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load your DataFrame (assuming df3 is your DataFrame)
# df3 = pd.read_csv('your_data_file.csv')  # Uncomment if needed

# Features to normalize
features_to_normalize = [
    'Total_Athletes', 'Events_Count', 'population_total', 
    'GNI_per_capita', 'GDP', 'GDP_per_capita', 
    'population_growth', 'GDP_growth'
]

# Initialize the scaler
scaler = StandardScaler()

# Apply Z-score normalization
df5[features_to_normalize] = scaler.fit_transform(df5[features_to_normalize])

# Verify the normalization
print(df5[features_to_normalize].head())


```

```python
# Apply one-hot encoding to the 'Region' column
df5_encoded = pd.get_dummies(df5, columns=['Region'], drop_first=True)

# Verify the encoding
print(df5_encoded.head())
```

## ML Model 

```python
# Assuming 'Year' is in your DataFrame and it's numeric or convertible
df_train = df5_encoded[df5_encoded['Year'].between(1992, 2016)]
df_test = df5_encoded[df5_encoded['Year'] == 2020]

# Features and target variable
X_train = df_train[['Total_Athletes', 'Events_Count', 'population_total', 'GNI_per_capita', 'GDP', 'GDP_per_capita', 
                    'population_growth', 'GDP_growth', 'Host_or_Not', 'HDI', 'Population_Percentage', 'GDP_Percentage'] + [col for col in df_train.columns if col.startswith('Region_')]]

y_train = df_train['Total_Medals']

X_test = df_test[['Total_Athletes', 'Events_Count', 'population_total', 'GNI_per_capita', 'GDP', 'GDP_per_capita', 
                  'population_growth', 'GDP_growth', 'Host_or_Not', 'HDI', 'Population_Percentage', 'GDP_Percentage'] + [col for col in df_test.columns if col.startswith('Region_')]]

y_test = df_test['Total_Medals']
```

### 1. Training the Linear Regression Model

```python

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
predictions = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')


```
