```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso, PoissonRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Assuming df1 is your DataFrame with the necessary data already loaded
# Prepare your training and test data
X_train = df1[df1['Year'] <= 2016].drop(['Total_Medals', 'Year'], axis=1)
y_train = df1[df1['Year'] <= 2016]['Total_Medals']

X_test = df1[df1['Year'] == 2020].drop(['Total_Medals', 'Year'], axis=1)
y_true = df1[df1['Year'] == 2020]['Total_Medals']
countries = df1[df1['Year'] == 2020]['Country']

X_train = X_train.drop(['Gold_Medals', 'Silver_Medals', 'Bronze_Medals','Country','Country_Code'], axis=1)
X_test = X_test.drop(['Gold_Medals', 'Silver_Medals', 'Bronze_Medals','Country','Country_Code'], axis=1)


# Define the feature columns
numerical_features = [
    'population_total', 'population_growth', 'GNI_per_capita', 
    'GDP', 'GDP_growth', 'GDP_per_capita', 
    'Population_Percentage', 'GDP_Percentage', 'HDI'
]
categorical_features = ['Region']

# Create a ColumnTransformer to handle both numerical and categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(drop='first'), categorical_features)
    ]
)

# Fit and transform the training data
X_train_processed = preprocessor.fit_transform(X_train)

# Transform the test data with the same preprocessor
X_test_processed = preprocessor.transform(X_test)

# Initialize models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'Poisson': PoissonRegressor(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Support Vector Regression': SVR(kernel='linear')
}

# Prepare a DataFrame to store results
results = []

# Train each model, make predictions and evaluate
for name, model in models.items():
    model.fit(X_train_processed, y_train)
    
    # Predictions
    y_pred_train = model.predict(X_train_processed)  # Predictions on training set
    y_pred_test = model.predict(X_test_processed)    # Predictions on test set
    
    # Calculate metrics
    train_mse = mean_squared_error(y_train, y_pred_train)  # MSE on training set
    test_r2 = r2_score(y_true, y_pred_test)                # R^2 on test set
    
    # Store results
    results.append({
        'Model': name,
        'Train_MSE': train_mse,
        'Test_R2': test_r2
    })

# Create a DataFrame with results
results_df = pd.DataFrame(results)

# Display the results
print(results_df)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso, PoissonRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Assuming df1 is your DataFrame with the necessary data already loaded
# Prepare your training and test data
X_train = df1[df1['Year'] <= 2016].drop(['Total_Medals', 'Year'], axis=1)
y_train = df1[df1['Year'] <= 2016]['Total_Medals']

X_test = df1[df1['Year'] == 2020].drop(['Total_Medals', 'Year'], axis=1)
y_true = df1[df1['Year'] == 2020]['Total_Medals']
countries = df1[df1['Year'] == 2020]['Country']

X_train = X_train.drop(['Gold_Medals', 'Silver_Medals', 'Bronze_Medals', 'Country', 'Country_Code'], axis=1)
X_test = X_test.drop(['Gold_Medals', 'Silver_Medals', 'Bronze_Medals', 'Country', 'Country_Code'], axis=1)

# Define the feature columns
numerical_features = [
    'population_total', 'population_growth', 'GNI_per_capita', 
    'GDP', 'GDP_growth', 'GDP_per_capita', 
    'Population_Percentage', 'GDP_Percentage', 'HDI'
]
categorical_features = ['Region']

# Create a ColumnTransformer to handle both numerical and categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(drop='first'), categorical_features)
    ]
)

# Fit and transform the training data
X_train_processed = preprocessor.fit_transform(X_train)

# Transform the test data with the same preprocessor
X_test_processed = preprocessor.transform(X_test)

# Initialize models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'Poisson': PoissonRegressor(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Support Vector Regression': SVR(kernel='linear')
}

# Prepare a list to store results
results = []

# Train each model, make predictions, and evaluate
for name, model in models.items():
    model.fit(X_train_processed, y_train)
    
    # Predictions
    y_pred_train = model.predict(X_train_processed)  # Predictions on the training set
    y_pred_test = model.predict(X_test_processed)    # Predictions on the test set
    
    # Calculate metrics
    train_mse = mean_squared_error(y_train, y_pred_train)  # MSE on training set
    test_r2 = r2_score(y_true, y_pred_test)                # R^2 on the test set
                       
    std_pred_test = y_pred_test.std()                      # Standard deviation of test set predictions
    
    # Store results
    results.append({
        'Model': name,
        'Train_MSE': train_mse,
        'Test_R2': test_r2,
        
        'Std_Dev_Predicted_Test': std_pred_test
    })

# Create a DataFrame with results
results_df = pd.DataFrame(results)

# Display the results
print(results_df)
```
